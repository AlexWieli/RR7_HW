### *Show description and summary*

**Game of Thrones** is an American fantasy drama television series created by David Benioff and D. B. Weiss for HBO. It is an adaptation of A Song of Ice and Fire, a series of fantasy novels by George R. R. Martin, the first of which is **A Game of Thrones**. The show premiered on HBO in the United States on April 17, 2011, and concluded on May 19, 2019, with 73 episodes broadcast over eight seasons.

Set on the fictional continents of Westeros and Essos, **Game of Thrones** has a large ensemble cast and follows several story arcs throughout the course of the show. The first major arc concerns the Iron Throne of the Seven Kingdoms of Westeros through a web of political conflicts among the noble families either vying to claim the throne or fighting for independence from whoever sits on it. The second major arc focuses on the last descendant of the realm's deposed ruling dynasty, who has been exiled to Essos and is plotting to return and reclaim the throne. The third follows the Night's Watch, a military order defending the realm against threats from beyond Westeros' northern border.

**Game of Thrones** attracted a record viewership on HBO and has a broad, active, and international fan base. Critics have praised the series for its acting, complex characters, story, scope, and production values, although its frequent use of nudity and violence (including sexual violence) generated controversy. The final season received significant criticism for its reduced length and creative decisions, with many considering it a disappointing conclusion. The series received 59 Primetime Emmy Awards, the most by a drama series, including Outstanding Drama Series in 2015, 2016, 2018 and 2019. Its other awards and nominations include three Hugo Awards for Best Dramatic Presentation, a Peabody Award, and five nominations for the Golden Globe Award for Best Television Series – Drama.

### *Viewers per episode released*

Below a summary of **Game of Thrones** viewership over the episodes may be seen.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

views = pd.read_excel('Viewership.xlsx')

sns.set_palette("Set2")

plt.figure(figsize=(12, 6))

seasons = views['Season'].unique()

for season in seasons:
    season_data = views[views['Season'] == season]
    plt.plot(season_data['Episode'], season_data['View'], label=f'Season {season}')

plt.title('Average Viewers / Episode')
plt.xlabel('Episode')
plt.ylabel('Average Viewers')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

As can be seen on the graph, the show gained **heavy momentum** and **increased with popularity each season**. Season 5 appears to be the only one recieved poorly, which is indicated by a **heavier drop** in viewership. It appears that episodes 6 usually suffer a lower viewership, which could be an indicator of their **lower quality**. This cannot be said for the ending, as it's been seen by the most amount of viewers.
